```python
#CHECKING_DATA_QUALITY
#Check for Missing Values
missing_values = testing.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
def missing_values_table(testing):
  mis_val = testing.isnull().sum()
  mis_val_percent = 100*testing.isnull().sum() / len(testing)
  mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
  mis_val_table_ren_columns = mis_val_table.rename(columns = {0:'Missing Values' , 1:'% of Total Values'})
  return mis_val_table_ren_columns.round(1)
missing_values_table(testing)
```

```python
#Drop the null values
columns_to_drop_missingvalues = ['Loan ID', 'Customer ID', 'Current Loan Amount', 'Term',
                        'Home Ownership', 'Purpose', 'Monthly Debt', 'Years of Credit History',
                        'Number of Open Accounts', 'Number of Credit Problems',
                        'Current Credit Balance', 'Maximum Open Credit','Tax Liens']
testing.dropna(subset=columns_to_drop_missingvalues , inplace=True)
```

```python
#Dealing with null values in Credit Score
import matplotlib.pyplot as plt
credit = testing['Credit Score']
credit.plot.hist(bins=20,alpha=0.5)
plt.xlabel('Credit Score')
plt.title('Distribution of Credit Score')
plt.show()
```

Because its a positive skewed distribution, We will use the Mode strategy to deal with null values in Credit Score column 

```python
#Dealing with null values in Credit Score
from sklearn.impute import SimpleImputer
mode_imputer = SimpleImputer(strategy='most_frequent')
column_impute = 'Credit Score'
imputed_column = mode_imputer.fit_transform(testing[[column_impute]])
testing[column_impute] = imputed_column

#Check if it worked or not 
missing_credit_scores = testing['Credit Score'].isnull().sum()
print(missing_credit_scores)
```

```python
#Dealing with null values in Annual Income
import matplotlib.pyplot as plt
annual = testing['Annual Income']
annual.plot.hist(bins=20,alpha=0.5)
plt.xlabel('Annual Income')
plt.title('Distribution of Annual Income')
plt.show()
```

Because its a positive skewed distribution, We will use the Mode strategy to deal with null values in Annual Income column 

```python
#Dealing with null values in Annual Income
mode_imputer = SimpleImputer(strategy='most_frequent')
annual_income ='Annual Income'
imputed_annual = mode_imputer.fit_transform(testing[[annual_income]])
testing[annual_income] = imputed_annual

#Check if it worked or not 
missing_annual_income = testing['Annual Income'].isnull().sum()
print(missing_annual_income)
```

```python
#Dealing with null values in Years in current job
testing['Years in current job'].value_counts().plot(kind='bar')
plt.xlabel('Years in Current Job')
plt.ylabel('Count')
plt.title('Distribution of Years in Current Job')
plt.show()
```

Because its a positive skewed distribution, We will use the Median strategy to deal with null values in Years in current job column 

```python
#Dealing with null values in Years in current job
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

#Since its a categorical column, we will encode 'Years_in_current_job' column
le = LabelEncoder()
testing['Years in current job'] = le.fit_transform(testing['Years in current job'])

median_imputer = SimpleImputer(strategy='median')
imputed_year = median_imputer.fit_transform(testing[['Years in current job']])
testing['Years in current job'] = imputed_year

#Check if it worked or not 
missing_year = testing['Years in current job'].isnull().sum()
print(missing_year)
```

```python
#Dealing with null values in Bankruptcies
testing['Bankruptcies'].value_counts().plot(kind='bar')
plt.xlabel('Bankruptcies')
plt.ylabel('Count')
plt.title('Distribution of Bankruptcies')
plt.show()
```

Because its a positive skewed distribution, We will use the Mode strategy to deal with null values in Bankruptcies column 

```python
#Dealing with null values in Bankruptcies
mode_imputer = SimpleImputer(strategy='most_frequent')
bankruptcies ='Bankruptcies'
imputed_bankruptcies = mode_imputer.fit_transform(testing[[bankruptcies]])
testing[bankruptcies] = imputed_bankruptcies

#Check if it worked or not 
missing_bankruptcies = testing['Bankruptcies'].isnull().sum()
print(missing_bankruptcies)
```

```python
#Drop the columns that have many missing values 
testing.drop(columns='Months since last delinquent', axis=1, inplace=True) 

#Check missing values again
missing_values_table(testing)
```
